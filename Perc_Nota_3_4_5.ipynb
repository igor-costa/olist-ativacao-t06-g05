```sql
-- Percentual de avaliações recebidas pelo vendedor com nota 3, 4 e 5 (D-28, D-56, D-365, lifetime)

WITH base AS (
    SELECT DISTINCT
        t1.seller_id,
        t2.order_id,
        t2.review_score,
        t3.order_purchase_timestamp
    FROM workspace.olist.order_items t1
    JOIN workspace.olist.order_reviews t2
        ON t1.order_id = t2.order_id
    JOIN workspace.olist.orders t3
        ON t1.order_id = t3.order_id
    WHERE t3.order_purchase_timestamp <= '2018-07-02'
)

SELECT 
    seller_id,

    -- D-28
    AVG(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' AND review_score = 3
             THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' THEN 0 END) AS perc_nota_3_d28,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' AND review_score = 4 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' THEN 0 END) AS perc_nota_4_d28,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' AND review_score = 5 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02' THEN 0 END) AS perc_nota_5_d28,

    -- D-56
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' AND review_score = 3 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' THEN 0 END) AS perc_nota_3_d56,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' AND review_score = 4 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' THEN 0 END) AS perc_nota_4_d56,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' AND review_score = 5 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02' THEN 0 END) AS perc_nota_5_d56,

    -- D-365
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' AND review_score = 3 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' THEN 0 END) AS perc_nota_3_d365,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' AND review_score = 4 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' THEN 0 END) AS perc_nota_4_d365,
    AVG(CASE WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' AND review_score = 5 THEN 1
             WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02' THEN 0 END) AS perc_nota_5_d365,

    -- Lifetime
    AVG(CASE WHEN review_score = 3 THEN 1 ELSE 0 END) AS perc_nota_3_lifetime,
    AVG(CASE WHEN review_score = 4 THEN 1 ELSE 0 END) AS perc_nota_4_lifetime,
    AVG(CASE WHEN review_score = 5 THEN 1 ELSE 0 END) AS perc_nota_5_lifetime

FROM base
GROUP BY seller_id;
```
