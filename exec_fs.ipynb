```python
def import_query(path):
    with open(path, 'r') as open_file:
        query = open_file.read()
    return query

query = import_query("fs_products.sql")
```

```python
import pandas as pd

dates = pd.date_range("2017-01-01", "2018-07-01", freq="MS")
dates = list(map( lambda x: x.strftime("%Y-%m-%d"), dates ))
dates
```

```python
from tqdm import tqdm

for d in tqdm(dates):
    
    spark.sql(f"DELETE FROM workspace.feature_store.fs_products WHERE dtRef = '{d}' ")

    df = spark.sql(query.format(date=d))
    
    (df.write
       .mode("append")
       .saveAsTable("workspace.feature_store.fs_products"))
    
```
